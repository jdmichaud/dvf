```python
import pandas as pd
import numpy as np
import scipy
```

```python
header = []
with open('valeursfoncieres-2014.txt') as vf:
    headerline = vf.readline()
    header = headerline.strip().split('|')
_ = [print(i, c) for (i, c) in enumerate(header)]
```

```python
vf = pd.read_csv('valeursfoncieres-2014.txt', sep='|')
vf['Valeur fonciere'] = vf['Valeur fonciere'].str.replace(',', '.')
vf['Valeur fonciere'] = pd.to_numeric(vf['Valeur fonciere'])
vf['Surface reelle bati'] = pd.to_numeric(vf['Surface reelle bati'])
vf = vf.dropna(subset=['Valeur fonciere', 'Surface reelle bati'])

apparts = vf[(vf['Type local'] == 'Appartement') & (vf['Nature mutation'] == 'Vente')].copy()
# Remove outliers on input data
apparts = apparts[(np.abs(scipy.stats.zscore(apparts['Surface reelle bati'])) < 3)]
apparts = apparts[(np.abs(scipy.stats.zscore(apparts['Valeur fonciere'])) < 3)]
# Remove outliers on output data
apparts['prixm2'] = apparts['Valeur fonciere'] / apparts['Surface reelle bati']
apparts = apparts[(np.abs(scipy.stats.zscore(apparts['prixm2'])) < 3)]
apparts['prixm2']

len(apparts)
```

```python
def display_info(column, name):
    mean = column.mean()
    median = column.median()
    print(f'{name} mean {mean}')
    print(f'{name} median {column.median()}')
scipy.stats.zscore(apparts['Surface reelle bati'])
display_info(apparts['Surface reelle bati'], 'surface')
display_info(apparts['Valeur fonciere'], 'valeur')
```
